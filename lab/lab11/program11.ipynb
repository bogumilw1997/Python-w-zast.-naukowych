```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_path = 'data/lastfm_asia_edges.csv'
df = pd.read_csv(data_path)
Graphtype = nx.Graph()
G = nx.from_pandas_edgelist(df,'node_1', 'node_2', edge_attr=None, create_using=Graphtype)
```

```python
print(f'Liczba wierzchołków: {G.number_of_nodes()}')
print(f'Liczba krawędzi: {G.number_of_edges()}')
print(f'Spójny: {nx.is_connected(G)}')
print(f'Liczba spójnych komponentów: {nx.number_connected_components(G)}')
print(f'Średnia długość najkrótszej ścieżki: {nx.average_shortest_path_length(G, weight=None, method=None)}')
```

```python
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
plt.rcParams["figure.figsize"] = [20, 10]
```

```python
plt.bar(*np.unique(degree_sequence, return_counts=True))
plt.title("Histogram stopni wierzchołków")
plt.xlabel("Stopień")
plt.ylabel("Liczba wierzchołków")
plt.xlim(0,80)
```

```python
plt.rcParams["figure.figsize"] = [20, 10]
nx.draw(G,pos=nx.spring_layout(G), node_size = 5, width = 0.1, alpha = 0.4)
```
