```python
from bokeh.plotting import figure,show, row, column
from bokeh.io import output_notebook
import numpy as np
output_notebook()
```

```python
x = np.linspace( - 20* np.pi, 20* np.pi, 1000)
y_sin = np.sin(x)
y_cos = np.cos(x)

data = {'x':x, 'y_sin':y_sin, 'y_cos':y_cos}
```

```python
fig = figure(x_range = (-np.pi, np.pi), x_axis_label = 'x', y_axis_label = 'y', title = 'Tytuł', sizing_mode = 'stretch_width', height = 200)
fig.toolbar.logo = None
fig.grid.grid_line_dash = (6,5)
fig.line(x, y_sin)
fig.scatter(x, y_sin, color = 'purple', radius = 0.1 * np.log(abs(x)), fill_alpha = 0.5)
show(fig)
```

```python
fig1 = figure(x_range = (-np.pi, np.pi), x_axis_label = 'x', y_axis_label = 'y', title = 'Tytuł', sizing_mode = 'stretch_width', height = 200)
fig1.toolbar.logo = None
fig1.grid.grid_line_dash = (6,5)
fig1.line(x, y_sin)

fig2 = figure(x_range = (-np.pi, np.pi), x_axis_label = 'x', y_axis_label = 'y', title = 'Tytuł', sizing_mode = 'stretch_width', height = 200)
fig2.toolbar.logo = None
fig2.grid.grid_line_dash = (6,5)
fig2.scatter(x, y_sin, color = 'purple', radius = 0.1 * np.log(abs(x)), fill_alpha = 0.5)

show(column(fig1, fig2), sizing_mode = 'stretch_width')
```
